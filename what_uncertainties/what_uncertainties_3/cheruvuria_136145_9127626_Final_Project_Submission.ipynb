## Exposition

**Introduction and Literature Review:**

Incorporating uncertainty into deep learning models has increasingly become important due to the challenges associated with validating the accuracy of the predictions of deep learning models.

Consider two types of tasks for Bayesian modeling - house price prediction and object classification - to motivate the need for this research. 

For both tasks, we can identify two types of uncertainity stemming from predictive uncertainity:

*   Aleatoric uncertainty, for capturing the natural stochasticity of observations (e.g. house market volatility) - aleatoric uncertainty cannot be reduced even when more data is provided. We can furthermore break down aleatoric uncertainity into:

  - Uncertainity for the variance is constant for all samples (e.g. measurement error, known as homoscedastic uncertainity). This type of uncertainity can be useful for applications such as multi-task learning, where multiple objectives need to be learned from a shared representation (source: https://alexgkendall.com/computer_vision/bayesian_deep_learning_for_safe_ai/). 

  - Uncertainity for which the variance is input data-dependent (known as heteroscedastic uncertainity). We can predict this type of uncertainity as model output.

*   Epistemic uncertainity, for capturing uncertainity due to the model parameters - epistemic uncertainity can be reduced if more data is collected. For example, a house price prediction model trained on data for condos in California from 2010-2011 might exhibit higher epistemic uncertainity when predicting house prices for log homes in Flagstaff in 2015. Including more data from a larger time range and for different regions would help explain away this uncertainity.

But, why do we want to decompose predictive uncertainty to epistemic uncertainty and aleatoric uncertainty? Decomposing predictive uncertainty allows us to understand the different facets of an input and provides us clue to improve predictions. Epistemic uncertainty is related the existence of outliers relative to the traning distribution. More data will help to reduce the epistemic uncertainty. However, more data does NOT help to reduce aleatoric uncertainty. Instead, to decrease aleatoric uncertainty, we need more knowledge of the unobserved variables by adding more feature or conducting more refined/precise measurements. 

We can address this intuition in terms of plots with the data and model predictions. For example, in the case of regression tasks (Fig.1), we see aleatoric uncertainty in dense areas of the data (we see a blow-up in this regions), and we see epistemic uncertainty in gaps in the data.

Now that we've established the two primary types of uncertainity, we need to consider what types of solutions we can utilize to account for these types of uncertainities.

For the purposes of the research paper we're addressing, the authors utilize Bayesian Neural Networks. The motivation for this is as follows:

The predictions using Deep Neural Network (DNN) and Bayesian Neural Network (BNN) are distinct: The former generates point estimates and the latter generates predictive distributions. Interestingly, the point estimates generated by DNNs is a form of aleatoric uncertainity, and the predictive distributions generated by BNNs is a form of epistemic uncertainity.

The uncertainty provided by predictive distributions (using BNN) can tell us if the noise in predictions is due to our ignorance of the original data or due to the existence of the unobserved variables our neural networks (or any other models) fail to capture. However, the point estimates (using DNN) are NOT able to provide the information of output uncertainties.

Extending from these points, it seems that epistemic uncertainity is more challenging to model. In the paper, the authors utilize a variation of Dropout (a node commonly used in DNNs for regularization), called MCDropout, to account for epistemic uncertainity in their models. We first explore this formulation as part of our literature review on the theoretical justification for this solution:

(1) **Formulations of Epistemic Uncertainty (Weight Uncertainty):** 

(1-1) **Paper "Dropout as a Bayesian Approximation: Representing Model Uncertainty in Deep Learning"** (http://proceedings.mlr.press/v48/gal16.pdf)

This paper showed that a neural network with dropout applied before every weight layer is mathematically equivalent to an approximation using variational inference to the probabilistic deep Gaussian process. Dropout was orignally used as a method to avoid overfitting. However, from a different perspective, this paper demonstated that dropout in neural networks can be interpreted as a Bayesian approximation of the Gaussian process. More specifically, to compute the posterior distribution, we can simply have a Bernoulli distribution over the weights, and then perform several rounds of forward passes through the neural network. After this, we average the weigths obtained after each round of foward pass, which is actually same as doing Monte Carlo integration to find the expected output value of the model under the posterior distribution.

(1-2) **Paper "Weight Uncertainty in Neural Networks"** (https://arxiv.org/pdf/1505.05424.pdf)

Let $D$ be training data and $W$ be weights in the BNN. Then posterior distribution is $P(W|D)$. Now we use a variational inference method to approximate the posterior distribution, through which we want to find the parameters $\theta$ of a distribution $q(w|\theta)$ that can minimize the KL-divergence with the true posterior: 
$$
\theta^{\star} = \arg \min _{\theta} KL[q(W|D) || P(W|D)] = \arg \min _{\theta} KL[q(W|D) || P(W)] - E_{q(W|\theta)} [\log P(D|W)]. 
$$

Next, we look at formulations of aleatoric uncertainity, which the authors implemented through learned loss attenuation - we will cover this particular concept in detail in later sections, interspersed with the code.

(2) **Formulations of Aleatoric Uncertainty (Data Uncertainty):** 

**Paper "Frequentist uncertainty estimates for deep learning"** (http://bayesiandeeplearning.org/2018/papers/31.pdf)

Quantile regression means using regression estimate a given quantile level $\tau$ of the target variable $Y$, when conditioned to the values $x$ taken by a feature variable $X$. Then this paper estimated all the quantile levels simultaneously by minimizing a function that is the expectation of the difference between the real quantile and predicted quantile. Then the resulting function will be used to compute the entire conditional distribution of the output and and the aleatoric uncertainty. 

(3) 
**Paper "Bayesian Uncertainty Estimation for Batch Normalized Deep Networks"** 
(https://arxiv.org/pdf/1802.06455.pdf)

This paper showed that batch normalization in neural networks is another approach for approximate Bayesian inference. The possible reason for this is that we have stochasticity involved in sampling mini-batches for getting batch statistics in batch normalization.  A network trained with Batch norm in all its layers is equivalent to the variational loss in a neural network. 

(4) **A Formulation of Predictive Uncertainty Decomposition:** we can express the predictive uncertainty as the entropy of softmax distribution (https://dl.acm.org/citation.cfm?id=3305504):
$$
H[\hat{y} | x,D],
$$
where $H$ is entropy, $x$ presents the input, $D$ represents the training set, and $\hat{y}$ represents the estimates of output $y$ given input $x$ and training set $D$. 
The entropy $H$ can be decomposed as follows:
$$
H[\hat{y} | x,D] = I[\hat{y}, w| x,D] + E_{w \sim p(w|D)} [ H[\hat{y} | x,w]],
$$
where $I$ represents information gain and $E$ represents expectation. 

In the information gain $I[\hat{y}, w| x,D]$, input $x$ and training set $D$ are given (and fixed). Therefore, it is used to measure the uncertainty in the model's weights $w$ (that is, the epistemic uncertainty). On the other hand, the average entropy $E_{w \sim p(w|D)} [ H[\hat{y} | x,w]]$ is used to measure the uncertainty in the model's weights $w$ -- because the model's weights have been given (and fixed) and therefore this part of predictive uncertainty comes from the input $x$ rather than weights $w$.

## Helper functions for Regression Task

```python
#Code for metrics:

def recall_m(y_true, y_pred):
        true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
        possible_positives = K.sum(K.round(K.clip(y_true, 0, 1)))
        recall = true_positives / (possible_positives + K.epsilon())
        return recall

def precision_m(y_true, y_pred):
        true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
        predicted_positives = K.sum(K.round(K.clip(y_pred, 0, 1)))
        precision = true_positives / (predicted_positives + K.epsilon())
        return precision

def f1_m(y_true, y_pred):
    precision = precision_m(y_true, y_pred)
    recall = recall_m(y_true, y_pred)
    return 2*((precision*recall)/(precision+recall+K.epsilon()))

class Metrics(Callback):
  def on_train_begin(self, logs={}):
    self.val_f1s = []
    self.val_recalls = []
    self.val_precisions = []
    
  def on_epoch_end(self, epoch, logs={}):
    val_predict = (np.asarray(self.model.predict(self.validation_data[0]))).round()
    val_targ = self.validation_data[1]
    val_f1 = f1_score(val_targ, val_predict, average='micro')
    val_recall = recall_score(val_targ, val_predict, average='micro')
    val_precision = precision_score(val_targ, val_predict, average='micro')
    self.val_f1s.append(val_f1)
    self.val_recalls.append(val_recall)
    self.val_precisions.append(val_precision)
    return
 
metrics = Metrics()
```

### Predicting Epistemic Uncertainty

The authors of the paper 'What Uncertainties Do We Need in Bayesian Deep
Learning for Computer Vision?' prefer Monte Carlo Dropout for approximate variational inference for complex computer vision models. They do this by training a model with dropout before every weight layer, and by also performing dropout at test time to sample from the approximate posterior. The authors claim that dropout can be interpreted as a variational Bayesian approximation, where the approximating distribution is a mixture of two Gaussians with small variances and the mean of one of
the Gaussians is fixed at zero.

We implement a neural network with 4 layers (5, 10, 20, 10, 5 nodes), and add a dropout layer between any two weight layers. The dropout nodes have a 10% probability of turning off its output at both training and testing stage. This makes the network output stochastic, and the whole model can be viewed as an ensemble of all model configurations that can be generated by this setup during inference.

The weights are regularized to prevent very large activation, which the network could try during training to overcome random dropout. We use mean squared error with L2 regularization of weights, as described in the paper:
$$L(\theta, p) = \frac{(y-\hat{y})^T(y-\hat{y})}{2N} + \frac{1-p}{2N} ||\theta||^2_2 \tag{1}$$
where $p$ is the dropout probability.

The epistemic uncertainty is evaluated by sampling from the MC dropout enabled neural network $T$ times and estimating the variance of the predicted output. Similarly, the expected value of true output $y$ is estimated by taking the average of model outputs.
$$E(y) = \frac{\sum_{t=1}^T \hat{y}}{T}$$
$$Var(y) = \frac{\sum_{t=1}^T \hat{y}(x)^T\hat{y}(x)}{T} - E(y)^TE(y)$$

```python
# Function to define a neural network with given specification, apply dropout and regularization
def architecture(model_type, n_hidden, input_dim, output_dim, dropout_prob, reg, act='tanh', verbose=True):
    inputs = Input(shape=(input_dim,))
    inter = Dense(n_hidden[0], activation=act,
                  kernel_regularizer=l2(reg))(inputs)
    for i in range(len(n_hidden) - 1):
        if model_type == "mcDropout":
          inter = Dropout(dropout_prob)(inter, training=True)
        #else:
        #  inter = Dropout(dropout_prob)(inter)
        inter = Dense(n_hidden[i+1], activation=act,kernel_regularizer=l2(reg))(inter)
    if model_type == "mcDropout":
      inter = Dropout(dropout_prob)(inter, training=True)
    #else:
    #  inter = Dropout(dropout_prob)(inter)
    outputs = Dense(output_dim, kernel_regularizer=l2(reg))(inter) 
    model = Model(inputs, outputs)
    if verbose:
      model.summary()
    return model
```

```python
# Function to generate and run the neural network with given data
def model_runner(X_train, y_train, MC_dropout=True, 
                dropout_prob=0.20, n_epochs=100, batch_size=500, 
                n_hidden=[100,100], act='relu', loss='mse', verbose=True):
  
    input_dim = X_train.shape[1]
    output_dim = y_train.shape[1]
    N = X_train.shape[0]
    reg = (1 - dropout_prob) / (2. * N) if dropout_prob > 0 else 0    # adding L2 regularization

    if MC_dropout is True:
      if verbose:
        print('McDropout NN fit')
      model_mc_dropout = architecture(model_type = 'mcDropout', 
                                      n_hidden=n_hidden, input_dim=input_dim, 
                                      output_dim=output_dim, 
                                      dropout_prob=dropout_prob, reg=reg, act=act, verbose=verbose)
    else:
      model_mc_dropout = architecture(model_type = 'original_model', 
                                      n_hidden=n_hidden, input_dim=input_dim, 
                                      output_dim=output_dim, 
                                      dropout_prob=dropout_prob, reg=reg, act=act, verbose=verbose)
    

    model_mc_dropout.compile(optimizer='rmsprop', loss=loss, metrics=['mae'])

    start_time = time.time()
    model_mc_dropout.fit(X_train, y_train, batch_size=batch_size, nb_epoch=n_epochs, verbose=verbose)
    training_time = time.time() - start_time

    return model_mc_dropout, training_time
```

```python
# Function to predict on the test dataset, and the epistemic uncertainty
def predictor(model_mc_dropout, 
              X_test=x_test, T=100):
    probs_mc_dropout = []
    for _ in range(T):
        probs_mc_dropout += [model_mc_dropout.predict(X_test,verbose=0)]
    predictive_mean = np.mean(probs_mc_dropout, axis=0)
    predictive_variance = np.var(probs_mc_dropout, axis=0)
    return probs_mc_dropout
```

**Effect of model complexity on Epistermic uncertainty**

In this section, we build multiple networks with different depths (number of layers) and widths (number of neurons per layer). Each network is trained with and without adding a dropout layer between its weights layers. 
The models are evaluated on a test data set using dropout layers again in order to simulate a Gaussian process. We obtain each time, a range of output values for each input scalar. This allows us to compute the standard deviation of the posterior distribution. When looking at all input values, we add up the standard deviations for each area of our data [:-3], [-3:-2], [-2:2], [2:3] and [3:]. We also compute the variance of the standard deviations for each area. This allows us to plot how epistemic uncertainty varies with network compexity and input data.

```python
from sklearn.metrics import r2_score

x_train_reshaped = x_train.reshape(x_train.shape[0], 1)
y_train_reshaped =  y_train.reshape(x_train.shape[0], 1)

# architecture space
nb_layers_min = 5
nb_layers_max = 50
nb_neurons_min = 5
nb_neurons_max = 100

x_test = np.linspace(-10,10,100)
y_test = x_test

# areas of data
x_1, x_2, x_3, x_4 = np.where(np.rint(x_test)==-3)[0][0], \
                     np.where(np.rint(x_test)==-2)[0][0], \
                     np.where(np.rint(x_test)==2)[0][0], \
                     np.where(np.rint(x_test)==3)[0][0]

# function to build a network with given layout, with or without dropout
# gets posterior distribution of model output on test dataset
# computes std and mean of predictions for each data area
# plots predictions with error bars
def get_variance(layout, ax, MC_dropout=True, n_epochs=10, batch_size=100, dropout_prob=0.05, act='relu'):
  model, training_time = model_runner(x_train_reshaped, y_train_reshaped, 
                                      MC_dropout=MC_dropout, 
                                      n_epochs=n_epochs, 
                                      batch_size=batch_size, 
                                      n_hidden=layout, 
                                      dropout_prob=dropout_prob, act=act, verbose=False)
  p = predictor(model, x_test, T=50)
  mean, std, var = np.mean(np.array(p), axis=0), np.std(np.array(p), axis=0), np.var(np.array(p), axis=0)
  std_total = [np.sum(std[:x_1]), np.sum(std[x_1:x_2]), np.sum(std[x_2:x_3]), np.sum(std[x_3:x_4]), np.sum(std[x_4:])]
  std_var = [np.var(std[:x_1]), np.var(std[x_1:x_2]), np.var(std[x_2:x_3]), np.var(std[x_3:x_4]), np.var(std[x_4:])]
  r2 = r2_score(y_train_reshaped, mean)
  ax.plot(x_test, y_test, ls='--')
  ax.scatter(x_train, y_train, color='black')
  ax.errorbar(x_test, mean, yerr=std, fmt='.')
  ax.set_title('layout {}x{} {} - R2 {:.2f}'.format(len(layout), layout[0], ('with dropout' if MC_dropout else 'without dropout'), 
                                          r2_score(y_test, mean)))
  ax.set_xlabel('x')
  ax.set_ylabel('y')
  return std_total, std_var, r2, model
```

### Predicting Aleatoric Uncertainty

The authors of the paper propose a modification to (1) to include direct prediction of heteroskedastic aleatoric uncertainty in the model.
$$L(\theta, p) = \frac{\sum_{i=1}^N (y_i-\hat{y}_i)^2}{2N\sigma(x_i)^2} + \frac{1}{2} \log \sigma(x_i)^2 \tag{2}$$

The inclusion of the term for aleatoric variance $\sigma(x_i)$ in (2) works because the model is now incentivized to predict higher aleatoric uncertainty in regions where it sees a higher error occuring. In order to prevent the model from predicting arbitrarily large variance everywhere, we regularize it by adding the second term $\log \sigma(x_i)^2$ to (2)

**Results for model with aleatoric uncertainity:**

We implement the above loss function below (see function 'aleatoric_loss') and fit the same neural network as before on the toy regression data. The model predicts both target $\hat{y}$ and aleatoric uncertainty $\log \sigma(x_i)^2$. The loss function used in the implementation is as follows:
$$L(\theta, p) = \frac{\sum_{i=1}^N \exp(-s_i)(y_i-\hat{y}_i)^2}{2N} + \frac{1}{2} s_i \tag{3}$$
where $s_i=\log \sigma_i^2$


```python
x_train = x_train.reshape(x_train.shape[0], 1)
y_train = np.vstack([y_train, np.zeros(y_train.shape)]).T
```

```python
def aleatoric_loss(y_true, y_pred):
    N = y_true.shape[0]
    se = K.pow((y_true[:,0]-y_pred[:,0]),2)
    inv_std = K.exp(-y_pred[:,1])
    loss1 = K.mean(inv_std*se)
    loss2 = K.mean(y_pred[:,1])
    return 0.5*(loss1 + loss2)

m_aleatoric, aleatoric_training_time = model_runner(x_train, y_train, n_epochs=1000, batch_size=10, n_hidden=[5,10,20,10,5], dropout_prob=0, act='relu', loss=aleatoric_loss)
m_total, total_training_time = model_runner(x_train, y_train, n_epochs=1000, batch_size=10, n_hidden=[5,10,20,10,5], dropout_prob=0.05, act='relu', loss=aleatoric_loss)
```

We plot the predicted $\hat{y}$ and the aleatoric variance below in the form of +/- 1 standard deviation error bars. We find that the aleatoric uncertainty is higher for data on the left than on the right as expected.

```python
# Aleatoric uncertainty
x_test = np.linspace(-10,10,100)
y_test = x_test
p = np.array(predictor(m_aleatoric, x_test, T=1))
mean, epistemic_std = np.mean(p[:,:,0], axis=0), np.std(p[:,:,0], axis=0)
aleatoric_std = np.exp(0.5*np.mean(p[:,:,1], axis=0))

plt.figure()
_=plt.plot(x_test, y_test, ls='--')
_=plt.scatter(x_train, y_train[:,0], color='black')
_=plt.errorbar(x_test, mean, yerr=aleatoric_std, fmt='.')
_=plt.title('Prediction with Aleatoric std')
```

Applying the modified loss function (2) on a neural network with Monte Carlo dropout applied, as in (1), we can predict both epistemic and aleatoric uncertainty from a single model.
$$L(\theta, p) = \frac{\sum_{i=1}^N (y_i-\hat{y}_i)^2}{2N\sigma(x_i)^2} + \frac{1}{2} \log \sigma(x_i)^2 + \frac{1-p}{2N} ||\theta||^2_2 \tag{4}$$

## Classification

### Explanation

**Epistemic** **Uncertainty**

For this project, we will use monte carlo dropout to estimate epistemic uncertainty for a classification problem using a simple classification toy dataset, MNIST dataset, and CIFAR dataset with a dropout probability of 20%. 

The loss function is obtained by calculating the categorical cross entropy of the prediction vector of the model, that was produced by applying softmax on the output of the model $x_{i,c}$.

$$y^{pred}_i = Softmax(x_{i,c}) $$

Loss function can thus be written as:
\begin{align}
L &= \sum_c y^{true}_{i} \log  y^{pred}_{i} \\
&=\sum_c y^{true}_{i} \log Softmax(x_{i,c}) \\
&= \sum_c y^{true}_{i} \log e^{x_{i,c} - \log \sum_c e^{x_{i,c}}}  
\end{align}

We use a numerically stable version of the softmax function, as suggested in the paper.
\begin{align}
Softmax(x_{i,c})&=\frac{e^{x_{i,c}}}{\sum_c e^{x_{i,c}}} \\
\log Softmax(x_{i,c}) &= x_{i,c} - \log \sum_c e^{x_{i,c}} \\
Softmax(x_{i,c}) &= e^{x_{i,c} - \log \sum_c e^{x_{i,c}}}
\end{align}



The prediction vector is obtained by averaging the sampled outputs from the MC dropout enabled model, while epistemic uncertainty is estimated by taking it's variance.
$$E(y^{pred}_i) = \frac{\sum_{t=1}^T y^{pred}_{i,t}}{T}$$
$$Var(y^{pred}_i) = \frac{\sum_{t=1}^T (y^{pred}_{i,t})^T y^{pred}_{i,t}}{T} - E(y^{pred}_{i,t})^TE(y^{pred}_{i,t}) \tag{5}$$



We apply a L2 regularization on the weights to prevent the activation function from blowing up. So the final loss function can be written as:
$$L = \sum_c y^{true}_{i} \log e^{x_{i,c} - \log \sum_c e^{x_{i,c}}}  + \frac{1-p}{2N} ||W||^2_2 \tag{6}$$

**Combining aleatoric and epistemic loss function**

The authors in the paper suggest that for classification problems, we can let the model predict aleatoric variance directly. The combined loss function for aleatoric and epistemic uncertainty can be derived as follows:

1. We take the output of the neural network model $W$ for each pixel $i$ and perturb it by some noise with predicted variance $\sigma^W_i$ T times ($t$ is for each iteration number)
$$x_{i,t}=f^W_i + \sigma^W_i \epsilon_t$$

2. We produce the predicted probability vector (Cx1) for that pixel by applying the softmax function.
\begin{align}
Softmax(x_{i,c,t}) &= \frac{e^{x_{i,c,t}}}{\sum_c e^{x_{i,c,t}}} \tag{7} \\
\log Softmax(x_{i,c,t}) &= x_{i,c,t} - \log \sum_c e^{x_{i,c,t}} \\
Softmax(x_{i,c,t}) &= e^{x_{i,c,t} - \log \sum_c e^{x_{i,c,t}}} \tag{8} \\
\end{align}

(7) is numerically more stable than (8), which is why we do it.

3. The above predicted probability vector is averaged T times.
$$y^{pred}_i = E_{\mathcal{N}}(Softmax(x_{i,c,t})) = \frac{1}{T}\sum_t e^{x_{i,c,t} - \log \sum_c e^{x_{i,c,t}}}$$

4. With this predicted probability vector $y^{pred}_i$ (Cx1), we compare it with the true probability vector $y^{true}_i$ and calculate the loss function as the  cross entropy:
\begin{align}
L_{x,\sigma^W, i} &= -\sum_c y^{true}_{i} \log  y^{pred}_{i} \\
&=-\sum_c y^{true}_{i} \log E_{\mathcal{N}}(Softmax(x_{i,c,t})) \\
&= -\sum_c y^{true}_{i} \log \frac{1}{T}\sum_t e^{x_{i,c,t} - \log \sum_c e^{x_{i,c,t}}}  \tag{9}
\end{align}

We also need to add a L2 regularization term to the loss function to prevent high activation due to dropout. In short the combined loss function can be concisely written as:
\begin{align}
x_{i,t}&=f^W_i + \sigma^W_i \epsilon_t \\
\epsilon_t &\sim \mathcal{N}(0,1) \\
L_{x,\sigma^W, i} &= -\sum_c y^{true}_{i} \log \frac{1}{T}\sum_t e^{x_{i,c,t} - \log \sum_c e^{x_{i,c,t}}}  + \frac{1-p}{2N} ||W||^2_2 \tag{10}
\end{align}

The above formulation (10) works as the model is incentivized to predict a high aleatoric variance $\sigma_i^W$ for regions where it makes large cross-entropy losses. By corrupting the model output with larger noise for such regions, the total cross entropy loss becomes restricted to around $\log 2$ (cross-entropy for random noise), which is much smaller than the loss that would have been incurred for a wrong prediction.

The aleatoric loss function is implemented below under the Python function 'bayesian_categorical_crossentropy'. We reuse the same neural network as for predicting epistemic uncertainty, except that we modify the loss function to (10). The output vector of the model is shaped 11x1 this time, with the additional prediction being for the log of aleatoric variance of the predicted vector. Epistemic variance is obtained directly by sampling and calculating the variance of the prediction vectors from the model multiple times. As MC dropout is retained in the model, outputs will have stochasticity in both training and testing phase.

```python
def softmax(pred):
  return K.exp(pred - K.log(K.sum(K.exp(pred))))  # numerically stable softmax
  
def softmax_np(pred):
  return np.exp(pred - np.log(np.sum(np.exp(pred), axis=1)).reshape(pred.shape[0],1))  

def gaussian_softmax(ypred, dist, num_classes):
  def map_fn(i):
    std_samples = K.transpose(dist.sample(num_classes))
    distorted_loss = softmax(ypred + std_samples)
    return distorted_loss
  return map_fn

def bayesian_categorical_crossentropy(T, num_classes):
  def bayesian_categorical_crossentropy_internal(ytrue, ypred_var):
    std = K.exp(0.5*ypred_var[:, num_classes:])
    ypred = ypred_var[:, 0:num_classes]
    iterable = K.variable(np.ones(T))
    dist = distributions.Normal(loc=K.zeros_like(std), scale=std)
    monte_carlo_results = K.map_fn(gaussian_softmax(ypred, dist, num_classes),
                                   iterable, name='monte_carlo_results')
    
    variance_loss = K.categorical_crossentropy(ytrue, K.mean(monte_carlo_results, axis=0))
    
    return variance_loss 
  
  return bayesian_categorical_crossentropy_internal

def simple_bayesian_categorical_crossentropy_internal(ytrue, ypred_var):
  #A slight modification to the original fuction for the simple example
    std = K.exp(0.5*ypred_var[:, num_classes:])
    ypred = ypred_var[:, 0:num_classes]
    iterable = K.variable(np.ones(T))
    dist = distributions.Normal(loc=K.zeros_like(std), scale=std)
    monte_carlo_results = K.map_fn(gaussian_softmax(ypred, dist, num_classes),
                                   iterable, name='monte_carlo_results')
    
    predictions = K.categorical_crossentropy(ytrue, y_pred)
    
    return predictions 
```

```python
def get_simple_model(act="relu", n_hidden=[32,64,64,32], dropout_prob=0.25, loss=keras.losses.categorical_crossentropy, include_logvar=False):
    N =  input_dimension [0]*input_dimension [1]
    reg = (1 - dropout_prob) / (2. * N) if dropout_prob > 0 else 0
    inputs = Input(input_shape)
    inter = Dense(n_hidden[0], activation=act,
                  kernel_regularizer=l2(reg))(inputs)
    for i in range(len(n_hidden) - 1):
        inter = Dropout(dropout_prob)(inter, training=True)
        inter = Dense(n_hidden[i+1], activation=act,
                      kernel_regularizer=l2(reg))(inter)
    inter = Dropout(dropout_prob)(inter, training=True)
    inter = Flatten()(inter)
    if include_logvar:
        outputs = Dense(num_classes+1, activation=None, kernel_regularizer=l2(reg))(inter) 
    else:
        outputs = Dense(num_classes, activation='softmax', kernel_regularizer=l2(reg))(inter) 
    
    model = Model(inputs=inputs, outputs=outputs)

    model.compile(loss=loss,
                  optimizer=keras.optimizers.Adadelta()) 
    return model
```

### MNIST - Classification - MC Dropout

MNIST data consists of images of handwritten digits from 0 to 9 that are prelabelled as such. The training dataset consists of 60,000 images while the test data contains 10,000 images. We use a similar neural network architecture as for regression, but with more nodes at each level (32,64,64 and 32 nodes). There output vector is of shape 10x1, for the 10 classes to be predicted. It is converted to a probability vector with the softmax function.
$$y^{pred}_i = Softmax(\hat{y})$$

```python
batch_size = 128
num_classes = 10
epochs = 12

# input image dimensions
img_rows, img_cols = 28, 28

# the data, split between train and test sets
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# add another dim to dataset for color - greyscale for MNIST would require 1 channel
x_train = x_train.reshape(x_train.shape[0], img_rows, img_cols, 1)
x_test = x_test.reshape(x_test.shape[0], img_rows, img_cols, 1)
input_shape = (img_rows, img_cols, 1)
print("Input Shape", input_shape)

x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train /= 255
x_test /= 255
print('x_train shape:', x_train.shape)
print(x_train.shape[0], 'train samples')
print(x_test.shape[0], 'test samples')

# convert class vectors to binary class matrices
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)
print(y_train.shape)
```

```python
def get_model(act="relu", n_hidden=[32,64,64,32], dropout_prob=0.25, loss=keras.losses.categorical_crossentropy, include_logvar=False):
    N =  input_shape[0]*input_shape[1]
    reg = (1 - dropout_prob) / (2. * N) if dropout_prob > 0 else 0
    inputs = Input(input_shape)
    inter = Dense(n_hidden[0], activation=act,
                  kernel_regularizer=l2(reg))(inputs)
    for i in range(len(n_hidden) - 1):
        inter = Dropout(dropout_prob)(inter, training=True)
        inter = Dense(n_hidden[i+1], activation=act,
                      kernel_regularizer=l2(reg))(inter)
    inter = Dropout(dropout_prob)(inter, training=True)
    inter = Flatten()(inter)
    if include_logvar:
        outputs = Dense(num_classes+1, activation=None, kernel_regularizer=l2(reg))(inter) 
    else:
        outputs = Dense(num_classes, activation='softmax', kernel_regularizer=l2(reg))(inter) 
    
    model = Model(inputs=inputs, outputs=outputs)

    metrics = ['accuracy'] if include_logvar else ['acc',f1_m,precision_m, recall_m]
    model.compile(loss=loss,
                  optimizer=keras.optimizers.Adadelta(),
                  metrics=metrics)
    return model
```

Below, we create the CNN model topology.

```python
def model_creator(act="relu", dropout_prob=0.25, loss=keras.losses.categorical_crossentropy, include_logvar=False):
  """
  Create CNN model using Keras Functional API.
  Architecture taken from https://keras.io/examples/cifar10_cnn/ and modified
  """
  input_shape = (32, 32, 3)
  N =  input_shape[0]*input_shape[1]
  reg = (1 - dropout_prob) / (2. * N) if dropout_prob > 0 else 0    # adding L2 regularization

  inputs = Input(shape=(32, 32, 3))
  inter = Conv2D(32, (3, 3), padding='same',
                  input_shape=x_train.shape[1:], kernel_regularizer=l2(reg))(inputs)
  inter = Activation('relu')(inter)
  inter = Conv2D(32, (3, 3), kernel_regularizer=l2(reg))(inter)
  inter = Activation('relu')(inter)
  inter = MaxPooling2D(pool_size=(2, 2))(inter)
  inter = Dropout(dropout_prob)(inter, training=True)
  inter = Conv2D(64, (3, 3), kernel_regularizer=l2(reg), padding='same')(inter)
  inter = Activation('relu')(inter)
  inter = Conv2D(64, (3, 3), kernel_regularizer=l2(reg))(inter)
  inter = Activation('relu')(inter)
  inter = MaxPooling2D(pool_size=(2, 2))(inter)
  inter = Dropout(dropout_prob)(inter, training=True)
  inter = Flatten()(inter)
  inter = Dense(512, kernel_regularizer=l2(reg))(inter)
  inter = Activation('relu')(inter)

  inter = Dropout(dropout_prob)(inter, training=True)
  inter = Dense(num_classes, kernel_regularizer=l2(reg))(inter)
  outputs = Activation('softmax')(inter)

  model = Model(inputs, outputs)
  metrics = ['accuracy'] if include_logvar else ['acc',f1_m,precision_m, recall_m]
  model.compile(loss=loss,
                  optimizer=keras.optimizers.RMSprop(lr=0.001, decay=1e-4), 
                  metrics=metrics)
  return model
```

## Conclusion

In this work, we exposed the scientific findings from the recently published paper “What Uncertainties Do We Need in Bayesian Deep Learning for Computer Vision?” by A Kendall. We introduced the notion of uncertainty in neural network and illustrated how to approach its quantification in a linear regression, polynomial regression and binary classification context. We further extended uncertainty evaluation when classifying images of handwritten digits using the well-known MNIST dataset. We also investigated uncertainty in classifying images of objects using the CIFAR-10 dataset.

Below we present an amalgamation of our findings:

Regression Task:
*   We implemented a neural network with a dropout layer between any two weight layers. This makes the network output stochastic. We trained the network on a synthetic linear regression dataset, where noise was introduced with a known magnitude. The epistemic uncertainty was evaluated by sampling from the MC dropout enabled neural network several times and estimating the variance of the predicted output. Aleatoric uncertainty was estimated by inclusion of a term for aleatoric variance in the loss function.
*   When varying model complexity on the same regression task, we found out that, using drop out, a low number of neurons is associated with higher epistemic uncertainty. Adding more layers did not seem to lower uncertainty significantly. Including epistemic uncertainty to the model by MC dropout does not seemed to achieve better regression.
*   Including support for both epistemic and aleatoric uncertainty provided a quantification of uncertainty that exactly followed our expectation. We were able to predict higher uncertainty for data portion where we included more noise. Epistemic uncertainty by MC dropout also seemed higher in regions where no training data was provided.

Simple Classification Example:
*   A toy classification problem consisting of two clusters of synthetic data was solved by fitting a network model with aleatoric and epistemic uncertainty. 
*   More work can be done to improve the decision boundary plots (which we were not able to do due to time constraints), but the general trend we noticed is that the epistemic uncertainity and combined models did seem to find the decision boundaries seperates the two clusters in our experiments.

MNIST Classification Tasks:
*   In this section, we studied the epistemic and aleatoric uncertainties in the classification of MNIST dataset. We tried three different types of images: (i) images with low uncertainty (that is, images that are clear); (ii) images with high uncertainty (that is, images that are vague); and (iii) random images(that is, images that are very vague). We found that the averaged prediction vector from the MC dropout enabled model has a higher accuracy than the best single prediction vector. Then we attempted different types of image augmentations (including rotation, vertical and horizontal shifts, and Laplacian filters) to see how these augmentaions influence the epistemic and aleatoric uncertainties.
